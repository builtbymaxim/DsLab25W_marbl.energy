# Setup & Data Loading (helper von GPT...)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from pathlib import Path

# Configuration
# We go up two levels (../../) to get to the root from notebooks/03_analysis/
PROFILES_DIR = Path('../../data/daily_profiles')
FIGURES_DIR = Path('../../data/figures')
FIGURES_DIR.mkdir(exist_ok=True, parents=True)

sns.set_style("whitegrid")

# Load the normalized profiles we created in Step 02
zones = ['DK1', 'ES', 'NO2']
data = {}

print("Loading Normalized Profiles...")
for zone in zones:
    file_path = PROFILES_DIR / f"{zone}_profiles.parquet"
    if file_path.exists():
        # Read parquet file
        df = pd.read_parquet(file_path)
        data[zone] = df
        print(f"  - {zone}: {len(df)} days loaded | Shape: {df.shape}")
    else:
        print(f"  ! {zone}: File not found at {file_path}")

print("\nReady for Cluster Analysis.")
```
